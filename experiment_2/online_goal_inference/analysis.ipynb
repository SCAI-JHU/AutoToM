```python
from scipy.stats import pearsonr

r_autotom, p_autotom = pearsonr(autotom_data.flatten(), human_data.flatten())
print(f'AutoToM and human: r={r_autotom:.4f}, p={p_autotom:.4f}')
r_gpt, p_gpt = pearsonr(gpt_data.flatten(), human_data.flatten())
print(f'GPT and human: r={r_gpt:.4f}, p={p_gpt:.4f}')
r_o3, p_o3 = pearsonr(o3_data.flatten(), human_data.flatten())
print(f'o3 and human: r={r_o3:.4f}, p={p_o3:.4f}')

def my_plot(m_data, h_data, save_name):
    plt.figure(figsize=(6, 6))
    plt.scatter(m_data.flatten(), h_data.flatten(), alpha=1.0)

    x = m_data.flatten()
    y = h_data.flatten()
    m, b = np.polyfit(x, y, 1)
    plt.plot(x, m*x + b, color='black', linestyle='-', linewidth=2.5)

    plt.ylim(0, 1)
    plt.xlim(0, 1)
    plt.grid(True, alpha=0.3)

    plt.xticks(np.arange(0, 1.1, 0.1), ['0', '', '', '0.3', '', '', '', '0.7', '', '', '1'], fontsize=25)
    plt.yticks(np.arange(0, 1.1, 0.1), ['0', '', '', '0.3', '', '','', '0.7', '', '',  '1'], fontsize=25)

    plt.tight_layout()
    plt.savefig(f'{save_name}_vs_human.pdf', dpi=300)

my_plot(autotom_data, human_data, 'autotom')
my_plot(gpt_data, human_data, 'gpt-4o')
my_plot(o3_data, human_data, 'o3-mini-high')
```
